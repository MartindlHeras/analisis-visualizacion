En primer lugar, leemos el archivo y creamos una columna extra para el mes y otra para el año:

```python
import pandas as pd

df = pd.read_csv('data_power_ree.csv')
df['Date'] = pd.to_datetime(df['Date'], utc=True)
df['Hora'] = df['Date'].dt.hour
df['Nombre Mes'] = df['Date'].dt.month_name('es_ES.utf8')
df['Mes'] = df['Date'].dt.month
df['Año'] = df['Date'].dt.year
df
```

```python
df.to_csv('data_power_ree_clean.csv')
```
